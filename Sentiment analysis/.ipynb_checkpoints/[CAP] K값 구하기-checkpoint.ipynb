```python
def totalWeight(comments_table):
    for i in range(len(comments_table)):
        totweight = comments_table.loc[i,'weight']*comments_table.loc[i,'vote']
        comments_table.loc[i,'totweight'] = totweight
    return comments_table
```

```python
def sentimentCount(comments_table):
    sentcnt = comments_table['sentiment'].value_counts()
    global p_count, n_count, e_count, com_count, p_vcount, n_vcount, e_vcount, com_vcount
    com_count = len(comments_table)
    com_vcount = comments_table['vote'].sum()
    
    for i in range(len(comments_table)):
        if comments_table.loc[i,'sentiment'] == 'Positive':
            p_vcount += comments_table.loc[i,'vote']
        elif comments_table.loc[i,'sentiment'] == 'Negative':
            n_vcount += comments_table.loc[i,'vote']
        else:
            e_vcount += comments_table.loc[i,'vote']
            
    for i in range(len(sentcnt)):
        if sentcnt.index[i] == 'Neutral':
            e_count = sentcnt[i]
        elif sentcnt.index[i] == 'Positive':
            p_count = sentcnt[i]
        else:
            n_count = sentcnt[i]
```

```python
def Kvalue(comments_table,com_vcount):    
    #K값 도출하기
    weight_sum = comments_table['totweight'].sum()
    global k_val
    Kweight = weight_sum/com_vcount
    k_val = 0.5-Kweight
```

```python
no_sent = 0
sentiment_table = pd.DataFrame(columns = ['date','p_count', 'n_count', 'e_count', 'com_count', 'p_vcount', 'n_vcount', 'e_vcount', 'com_vcount', 'k_val'])
sentiment_table.loc[:,'date'] = date_list
sentiment_table
```
